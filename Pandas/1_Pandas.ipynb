# Creating a Dataframe

```python
weather_data = {
    'day':['01/01/2022','01/02/2022','01/03/2022','01/04/2022','01/05/2022'],
    'temperature': [32,34,40,31,33],
    'windspeed': [6,7,8,5,6],
    'event': ['rainy', 'sunny', 'cloudy','rainy', 'sunny']
}

df = pd.DataFrame(weather_data)
df
```
